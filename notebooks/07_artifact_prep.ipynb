```python
import pandas as pd, joblib
from pathlib import Path

output_dir = Path("../data/outputs/")
X_train = pd.read_parquet(output_dir / "X_train.parquet")

joblib.dump(list(X_train.columns), "../deploy/artifacts/feature_cols.joblib")
joblib.dump(X_train.median().to_dict(), "../deploy/artifacts/feature_medians.joblib")


# Copy the 3 model artifacts
import shutil
shutil.copy(output_dir / "xgb_model.json", "../deploy/artifacts/xgb_model.json")
shutil.copy(output_dir / "nn_best.pt",     "../deploy/artifacts/nn_best.pt")
shutil.copy(output_dir / "scaler.joblib",  "../deploy/artifacts/scaler.joblib")

print(f"Feature count: {len(X_train.columns)}")  # confirm actual input_dim
```
