```python
import pandas as pd
```

```python
df=pd.read_stata('./data/raw/P2P_Macro_Data.dta')
```

```python
df.to_parquet('data/P2P_Macro_Data.parquet', index=False)
print("Saved as parquet")
```

```python
print("Shape:", df.shape)
print("\nbadloan distribution:")
print(df['badloan'].value_counts())
print(df['badloan'].value_counts(normalize=True).round(3))

print("\nMissing values (top 20):")
print(df.isnull().sum().sort_values(ascending=False).head(20))

print("\nDate range:")
print(df['issue_daily'].min(), "→", df['issue_daily'].max())
```

```python
fico_check = [c for c in df.columns if 'fico' in c.lower()]
print("FICO columns:", fico_check)
```

```python
# Check if baselinesample column exists
print('baselinesample' in df.columns)

# If it does:
if 'baselinesample' in df.columns:
    print(df['baselinesample'].value_counts())
    print("Baseline sample size:", df[df['baselinesample']==1].shape)
```

```python
#SANITY CHECK
# Check 1: does baselinesample exist?
print("baselinesample exists:", 'baselinesample' in df.columns)

# Check 2: default rate by year
print("\nDefault rate by year:")
print(df.groupby('issue_year')['badloan'].mean().round(3))

# Check 3: default rate by grade
print("\nDefault rate by grade:")
print(df.groupby('grade_')['badloan'].mean().sort_values(ascending=False).round(3))

# Check 4: pymnt_plan variance (we discussed this might be near-zero)
print("\npymnt_plan distribution:")
print(df['pymnt_plan'].value_counts())
```
